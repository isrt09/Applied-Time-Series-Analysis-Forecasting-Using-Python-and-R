```python
from dateutil.parser import parse
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('C:/Users/Mazedur Rahman/Desktop/monthly-sunspots.csv',parse_dates=['Month'])
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
# Time Plot
def plot_data(data, x,y, title='', xlabel='Month',ylabel='Sunspots', dpi=100):
    plt.figure(figsize=(16,2), dpi=dpi)
    plt.plot(x,y, color='tab:red')
    plt.gca().set(title=title,xlabel=xlabel,ylabel=ylabel)
    plt.show()
plot_data(data,x=data.index, y=data.Sunspots, title='Month vs Sunspots')
```

# Double Side Plot

```python
x  = data.index
y1 = data.Sunspots

plt.fill_between(x, y1=y1, y2=-y1, alpha=0.5, linewidth=2, color='skyblue')
plt.hlines(y=0, xmin = np.min(data.index), xmax = np.max(data.index), linewidth=5, color='red')
plt.show()
```
